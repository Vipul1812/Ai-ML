```python
def imageshow(image_path):
    img=cv2.imread(image_path)
    cv2.imshow(image_path,img)
    key=cv2.waitKey(0)

    if key == ord('q'):
        cv2.destroyAllWindows
```

```python
imageshow("AI-robo-lab.jpg")
```
