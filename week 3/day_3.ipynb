## DATA GROUPING

groupby()   function

```python
df.groupby('day')[['total_bill','tip']].mean()
```

```python
df.groupby('sex')[['total_bill','tip']].mean()
```

```python
df.groupby(['day','sex'])[['total_bill','tip']].mean()
```

## AGGRGATION

used for multiple grouping 

```python
import numpy as np
```

```python
df.groupby(['sex','day'])[['total_bill','tip']].agg([sum, np.mean ,np.std])
```

## FUNCTION

```python
df['total_price']=df['total_bill']+df['tip']
df.head()
```

```python
df.groupby('sex')['total_price'].mean()
```

```python
def mean_placer(data):
    return df[df['sex']==data['sex']]['total_price'].mean()
    


```

```python
df[['sex','total_price']].apply(mean_placer, axis=1 )
df.head()
```
