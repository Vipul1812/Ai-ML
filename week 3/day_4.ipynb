## DATA VISUALIZATION

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("tips.csv")
```

## LINE PLOT USING PANDA

```python
fig=plt.figure(figsize=(18,7))
df['total_bill'].plot(color='black', linestyle= '-', marker='p' , markerfacecolor='yellow')                             ## FIGURE


plt.title("line chart by panda")                    ## HEADING

plt.show()
```

## LINE PLOT USING SEABORN

```python
sns.lineplot(data=df, x=df.index, y='total_bill')

plt.title("line plot using seaborn")
plt.show()
```

## HISTOGRAM

```python
plt.title("histogram by using panda")
plt.subplot(1,2,1)

df['size'].plot.hist()

plt.grid()





plt.subplot(1,2,2)

sns.histplot(data=df,x='size' ,bins=5 , color='red')

plt.show()

```

## BOX PLOT

```python

fig=plt.figure(figsize=(18,7))
plt.subplot(1,2,1)
df['total_bill'].value_counts().plot.bar()
plt.title('panda using box plot')


plt.subplot(1,2,2)
sns.boxplot(data=df, x='day', y='total_bill')
plt.title('SEABORN USING BOX PLOT')
```
