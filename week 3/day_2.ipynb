## JOINING AND MERGING AND CONCATITION

```python
Df1=pd.DataFrame({'name':['Vipul','Bhumi','Pradeep','Sonu','shanu'],'id':[1,2,4,5,7]})
Df1
```

```python
df2=pd.DataFrame({'Domain':['IT','ARTS','Doctor','Web Dev','ittt'],'id':[1,2,4,5,6]})
df2
```

```python
df3=pd.DataFrame({'city':['ny','London','bikaner','torrato','delhi'],'id':[3,2,1,5,5]})
df3
```

merge

```python
pd.merge(df2,df3,on='id')   #common h vo aayegi
```

```python
pd.merge(df2,df3,on='id',how='left')        #Sab chaihiye tab how use kreyege
```

Concate 

```python
df44=pd.concat([Df1,df2])
df44   # horizontally               
```

```python
pd.concat([Df1,df2],axis=1)  #Vertically
```

Join

```python
Df1.join(df2,on='id',lsuffix='_left')
```

LOGICAL INDEXING

```python
df3[df3['city'].str.len()>5]
```
