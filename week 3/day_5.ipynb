## SCATTER PLOT

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("tips.csv")
```

```python
plt.figure(figsize=(18,6))
df.plot.scatter(x="total_bill",y="tip")
plt.title("Scatter plot using pandas")
plt.show()

sns.scatterplot(data=df,x='total_bill',y='tip')
plt.title("Scatter plot using Seaborn")
plt.show()
```

```python
sns.scatterplot(data=df,x='total_bill',y='tip', hue='sex', size='total_bill')
```

## HEXBIN

```python
df.plot.hexbin(x='total_bill',y='tip')
plt.title("hexavin Using Pandas")
plt.show()


sns.jointplot(data=df,x='total_bill',y='tip', kind='hex')
plt.show()
```

## KDE PLOT

```python
plt.subplot(121)

df['total_bill'].plot.kde()
plt.show()

plt.subplot(122)

sns.kdeplot(data=df, x='total_bill')
```

## PIE CHART


```python
df['day'].value_counts().plot.pie(autopct='%d')
```

## VIOLEN PLOT

```python
sns.violinplot(data=df,x='day' ,y='tip')
```

## SWARM PLOT

```python
sns.set_style('dark')
```

```python
sns.swarmplot(data=df,x='day' ,y='tip')
```

## HEAT MAP

```python
sns.heatmap(df[['total_bill','tip']].corr(),fmt='1,4f')
```
