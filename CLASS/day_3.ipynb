```python

import cv2
vid=cv2.VideoCapture(0)

# mai loop to read and show image until we bread the loop


while True:
        flag, img=vid.read()    #
        if flag:

            #DRAW THE RECTANGLE ON IMG
            cv2.rectangle(img,pt1=(10,10),pt2=(100,100),color=(255,0,0))
            cv2.imshow("Web_cam",img)   #showing by this
            key=cv2.waitKey(1)      # Wait for all sec
            if key== ord("q"):      #  ASCII OREDRED TABLE
                  break
        else:
             break
cv2.destroyAllWindows()     # Destroy The Image in This
vid.release
```
